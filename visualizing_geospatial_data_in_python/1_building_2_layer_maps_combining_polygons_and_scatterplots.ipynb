## Styling a scatterplot

In this exercise, you'll be using `plt.scatter()` to plot the father and son height data from the video. The `father_son` DataFrame is available. In each scatterplot, plot `father_son['fheight']` as x-axis and `father_son['sheight']` as y-axis.

Instructions

1. Import the `pyplot` module from `matplotlib` with the usual alias.
2. Create a scatterplot of father and son heights with a square marker (encoded as `s`) that is `'darkred'`. Show your plot.
3. Edit the code to change the markers for your scatterplot of father heights vs son heights. Make the markers `'yellow'` with a `'darkblue'` edgecolor.
4. Add gridlines and axes labels (`'father height (inches)'` and `'son height (inches)'`) to your scatterplot.
5. Give the plot a title of `'Son Height as a Function of Father Height'`.

```python
# Import pandas
import pandas as pd

# Import data
father_son = pd.read_csv('father_son.csv')
```

```python
# Import matplotlib.pyplot
import matplotlib.pyplot as plt

# Scatterplot 1 - father heights vs son heights with darkred square markers
plt.scatter(father_son['fheight'], father_son['sheight'], c='darkred', marker='s')

# Show your plot
plt.show()
```

```python
# Import matplotlib.pyplot
import matplotlib.pyplot as plt

# Scatterplot 2 - yellow markers with darkblue borders
plt.scatter(father_son['fheight'], father_son['sheight'], c='yellow', edgecolor='darkblue')

# Show the plot
plt.show()
```

```python
# Import matplotlib.pyplot
import matplotlib.pyplot as plt

# Scatterplot 3
plt.scatter(father_son['fheight'], father_son['sheight'], c='yellow', edgecolor='darkblue')
plt.grid()
plt.xlabel('father height (inches)')
plt.ylabel('son height (inches)')
plt.title('Son Height as a Function of Father Height')

# Show your plot
plt.show()
```

## Extracting longitude and latitude

A DataFrame named `df` has been loaded. Complete the code to extract longitude and latitude to new, _separate_ columns.

Instructions

1. Print the first few rows with `print()` and `df.head()`.
2. Use a list comprehension to extract latitude to a new column called `lat`.
3. Use a list comprehension to extract longitude to a new column called `lng`.
4. Print the first few rows with `print()` and `df.head()` again to confirm the new columns.

```python
# DataFrame df
df = pd.DataFrame([[4431, 'MUSIC CITY CENTRAL 5TH - BAY 11', (36.16659, -86.781996)],
                   [588, 'CHARLOTTE AVE & 7TH AVE N WB', (36.165, -86.78406)],
                   [590, 'CHARLOTTE AVE & 8TH AVE N WB', (36.164393, -86.785451)],
                   [541, '11TH AVE / N GULCH STATION OUTBOUND', (36.162249, -86.790464)]],
                   columns=['StopID', 'StopName', 'Location'])
```

```python
# print the first few rows of df 
display(df.head())

# extract latitude to a new column: lat
df['lat'] = [loc[0] for loc in df.Location]

# extract longitude to a new column: lng
df['lng'] = [loc[1] for loc in df.Location]

# print the first few rows of df again
display(df.head())
```
